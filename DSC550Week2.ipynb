### Introduction to the Dataset

The dataset I've chosen comprises two files related to the stock market over a 30-year period. The first file, 30_yr_stock_market_info.csv, contains information about different stock market symbols, including their names, types, full names, and units. The second file, 30_yr_stock_market_data.csv, is more extensive and provides daily data for various indices and commodities. 

This data includes indices like the Dow Jones, Nasdaq, S&P 500, NYSE Composite, and Russell 2000, as well as information on commodities like crude oil, copper, gold, and others. It also includes data on treasury yields of different durations.

#### Research Questions
Given the comprehensive nature of this dataset, I aim to explore the following questions:

#### 1. How have the major stock indices (Dow Jones, Nasdaq, S&P 500) performed over the last 30 years?

####  2. Is there a correlation between the performance of stock indices and commodities like gold and crude oil?

#### 3. How have economic indicators, such as treasury yields, fluctuated in relation to the stock market indices over this period?


### Graphical Analysis
To answer these questions, I will create and analyze three types of graphs:

1. Line Graphs for the major stock indices to observe their performance over time.

2. Scatter Plots to analyze the correlation between stock indices and commodities like gold and crude oil.

3. Line Graphs comparing the fluctuations in treasury yields with the stock market indices.


```python
import pandas as pd

# Load the dataset
file_path_1 = '/Users/bbeardsley/30_yr_stock_market_info.csv'
file_path_2 = '/Users/bbeardsley/30_yr_stock_market_data.csv'

# Inspecting the first few lines of each file to understand their structure and content
data_info = pd.read_csv(file_path_1)
data_stock = pd.read_csv(file_path_2)

data_info.head(), data_stock.head()
```

```python
import matplotlib.pyplot as plt

# Data preparation for the graphs
# Filtering out relevant columns for each analysis
stock_indices = data_stock[['Date', 'Dow Jones (^DJI)', 'Nasdaq (^IXIC)', 'S&P500 (^GSPC)']].copy()
stock_indices['Date'] = pd.to_datetime(stock_indices['Date'])

commodities = data_stock[['Date', 'Gold (GC=F)', 'Crude Oil-WTI (CL=F)']].copy()
commodities['Date'] = pd.to_datetime(commodities['Date'])

treasury_yields = data_stock[['Date', 'Treasury Yield 10 Years (^TNX)', 'Dow Jones (^DJI)']].copy()
treasury_yields['Date'] = pd.to_datetime(treasury_yields['Date'])

# Plotting the graphs
fig, axs = plt.subplots(3, 1, figsize=(15, 15))

# Graph 1: Line Graphs for Stock Indices
for column in ['Dow Jones (^DJI)', 'Nasdaq (^IXIC)', 'S&P500 (^GSPC)']:
    axs[0].plot(stock_indices['Date'], stock_indices[column], label=column)
axs[0].set_title('30-Year Performance of Major Stock Indices')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Index Value')
axs[0].legend()

# Graph 2: Scatter Plot for Commodities vs. Dow Jones
axs[1].scatter(commodities['Date'], commodities['Gold (GC=F)'], alpha=0.5, label='Gold Price')
axs[1].scatter(commodities['Date'], commodities['Crude Oil-WTI (CL=F)'], alpha=0.5, label='Crude Oil Price')
axs[1].set_title('Gold and Crude Oil Prices Over Time')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Price')
axs[1].legend()

# Graph 3: Line Graph for Treasury Yields vs. Dow Jones
axs[2].plot(treasury_yields['Date'], treasury_yields['Treasury Yield 10 Years (^TNX)'], label='10-Year Treasury Yield')
axs[2].plot(treasury_yields['Date'], treasury_yields['Dow Jones (^DJI)'], label='Dow Jones (^DJI)', alpha=0.5)
axs[2].set_title('Treasury Yield vs. Dow Jones Index Over Time')
axs[2].set_xlabel('Year')
axs[2].set_ylabel('Yield (%) / Index Value')
axs[2].legend()

plt.tight_layout()
plt.show()
```

### Analysis of the Graphs

#### 30-Year Performance of Major Stock Indices: 

The first graph shows a long-term upward trend in the major stock indices - Dow Jones, Nasdaq, and S&P 500. This growth is indicative of the overall strength and expansion of the stock market over the past three decades. However, there are periods of volatility and sharp declines, reflecting economic downturns, financial crises, and market corrections.


#### Gold and Crude Oil Prices Over Time: 

The scatter plot for gold and crude oil prices reveals fluctuations in commodity prices over the same period. Gold, often seen as a safe haven during economic uncertainty, shows price spikes during periods of market volatility. Crude oil prices also fluctuate significantly, influenced by global economic conditions, political events, and changes in supply and demand.


#### Treasury Yield vs. Dow Jones Index Over Time: 

The third graph compares the 10-year Treasury yield with the Dow Jones index. The yield often moves inversely to the stock market: when stock prices fall, treasury yields tend to decrease as investors move towards safer assets like government bonds, and vice versa. This inverse relationship is a classic indicator of investor sentiment and economic outlook.

#### Conclusion

The analysis of these graphs reveals key insights into the stock market, commodity prices, and economic indicators over the past 30 years:

The major stock indices have generally shown robust growth, punctuated by periods of significant volatility.
Commodity prices, particularly gold and crude oil, have fluctuated in response to various economic and geopolitical factors.


The inverse relationship between treasury yields and stock indices highlights the dynamic interplay between investor sentiment and economic conditions.
